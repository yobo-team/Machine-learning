**MODELING**

```python
books   = books[:10000]
```

```python
tfidf = TfidfVectorizer()
tfidf.fit(books.book_title)
# tfidf.get_feature_names_out()
```

```python
tfidf_matrix = tfidf.fit_transform(books.book_title)
tfidf_matrix.shape
```

```python
tfidf_matrix.todense()
```

```python
pd.DataFrame(
    tfidf_matrix.todense(),
    columns = tfidf.get_feature_names_out(),
    index   = books.book_title
).sample(20, axis=1).sample(10, axis=0)
```

```python
cosine_sim = cosine_similarity(tfidf_matrix)
cosine_sim
```

```python
cosine_sim_df = pd.DataFrame(
    cosine_sim,
    columns = books.book_title,
    index   = books.book_title
)

print(f'Cosine Similarity Shape : {cosine_sim_df.shape}')

cosine_sim_df.sample(8, axis=1).sample(8, axis=0)
```

```python
def author_recommendations_with_image(book_title, similarity_data=cosine_sim_df, items=books[['book_title', 'pub_year', 'image_l_url']], k=10):
    index = similarity_data.loc[:, book_title].to_numpy().argpartition(range(-1, -k, -1))
    closest = similarity_data.columns[index[-1:-(k+2):-1]]
    closest = closest.drop(book_title, errors='ignore')

    return pd.DataFrame(closest).merge(items).head(k)
```

```python
# Menghapus data yang terduplikasi berdasarkan 'book_title'
unique_books = books.drop_duplicates(subset=['book_title'])

# Mengurutkan DataFrame berdasarkan nama penulis (author) secara alfabetis A-Z
sorted_books = unique_books.sort_values(by='Book-Author', key=lambda x: x.str.lower())

# Mengambil 10 buku pertama
top_10_books = sorted_books.head(10)

# Menampilkan informasi 10 buku dengan ISBN, judul, tahun terbit, dan link gambar
print("Rekomendasi 10 Buku Berdasarkan Penulis (Author) A-Z:\n")
for i, (_, row) in enumerate(top_10_books[['isbn', 'book_title', 'pub_year', 'image_l_url']].iterrows(), start=1):
    isbn = row['isbn']
    book_title = row['book_title']
    pub_year = row['pub_year']
    image_url = row['image_l_url']

    print(f"{i}. ISBN: {isbn}")
    print(f"   Book Title: {book_title}")
    print(f"   Year: {pub_year}")
    print(f"   Image URL: {image_url}")
    print("\n")
```
